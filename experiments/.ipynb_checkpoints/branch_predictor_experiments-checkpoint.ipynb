# Gshare: misprediction rate plots for different traces and configs

```python
ghistorybits = [num_bits for num_bits in range(5, 16)]

int_1 = [35.57, 35.94, 34.23, 30.91, 26.11, 22.02, 19.09, 16.19, 13.84, 12.2, 11.2]
int_2 = [1.98, 1.71, 1.54, 1.26, 0.97, 0.73, 0.58, 0.50, 0.42, 0.401, 0.364]

mm_1 = [31.25, 29.89, 27.31, 22.5, 16.65, 13.1, 9.89, 7.88, 6.696, 5.33, 4.44]
mm_2 = [24.1, 21.41, 19.63, 17.63, 15.15, 13.31, 11.91, 10.93, 10.14, 9.09, 8.04]

fp_1 = [2.68, 2.31, 1.85, 1.65, 1.47, 1.22, 1.17, 1, 0.83, 0.83, 0.83]
fp_2 = [19.6, 17.64, 15.19, 11.45, 10.73, 6.13, 5.39, 3.36, 1.68, 1.36, 0.99]
```

```python
# make plot for gshare 
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(ghistorybits, int_1, c = 'b', label = "int_1")
plt.plot(ghistorybits, int_2, c = 'g', label = "int_2")
plt.plot(ghistorybits, mm_1, c = 'y', label = "mm_1")
plt.plot(ghistorybits, mm_2, c = 'orange', label = "mm_2")
plt.plot(ghistorybits, fp_1, c = 'r', label = "fp_1")
plt.plot(ghistorybits, fp_2, c = 'black', label = "fp_2")

plt.xlabel("#gHistoryBits")
plt.ylabel("Misprediction Rate (%)")
plt.title("Gshare: Misprediction rate against choice of #gHistoryBits")
plt.legend()
plt.show()
```
